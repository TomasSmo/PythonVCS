# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = list(range(100))
y = [i * 2 for i in x]
y_2 = [i ** 2 for i in x]
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax_2 = fig.add_axes([0.2, 0.5, .2, .2])
ax.plot(x, y)
ax_2.plot(x, y_2)
ax.set_xlabel("X'as")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax_2 = fig.add_axes([0.2, 0.5, .3, .3])
ax.plot(x, y)
ax_2.plot(x, y_2)
ax_2.set_xlim([40, 60])
ax_2.set_ylim([1000, 4000])
ax.set_xlabel("X'as")
plt.show()
```

```python
fig, ax = plt.subplots(2, 3)
#ax[0, 1].plot(x, y)
for row in ax:
    for col in row:
        col.plot(x,y)
```

# Seaborn


```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
ax = sns.countplot(data=tips, x='size', hue='sex', palette='vlag')
```

```python
sns.jointplot(data=tips, x='size', y='total_bill', kind='kde')
```

```python
sns.boxplot(data=tips, x='smoker', y='tip')
```

```python
titanic =sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.jointplot(data=titanic, x='fare', y='age', hue='class')
```

```python
sns.swarmplot(data=titanic, x='class', y='age', size=2, hue='sex', dodge=True)
```

```python
corr = titanic.corr()
sns.heatmap(corr, annot=True)
```

# Recursion

```python
def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))
```

```python
recur_fibo(20)
```
